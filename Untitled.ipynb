# Image Enhancement
#### Image enhancement module contains various classes which will used for image enhancement 
#### Classes are as follows:
#### Brightness()
#### color()
#### Contrast()
#### Sharpness()



```python
from PIL import Image
from PIL import ImageEnhance
image=Image.open("C://Users//PC//Desktop//C++ Code//xray.jpg")
resize=image.resize((256,256))

# Enhance Brightness
curr_img=ImageEnhance.Brightness(resize)
new=2.6
# Enhance Brightness by 2.6
Image_Brightened=curr_img.enhance(new)
display(curr_img,Image_Brightened)
```

### Color class

```python
## Enhance Color Level
curr_col=ImageEnhance.Color(resize)
## color level is enhanced by the factor of 2
img_color=curr_col.enhance(2)
display(img_color)
```

### contrast of image

```python
## contrast image

cotrast_image=ImageEnhance.Contrast(resize)

## contrast the image by the factor of 2
img=cotrast_image.enhance(2)
display(img)
```

### Sharpness of image

```python
Sharp_image=ImageEnhance.Sharpness(resize)
img=Sharp_image.enhance(2)
display(img)
```
